```python
## Read Binary stack and rewrites along a new axis
def BinaryRead(stack_directory):
    #Generate
    img_bin = np.zeros((901,1024,1536))
    i=0
    for i in list(range(900)):
    #for name in sorted(os.listdir(stack_directory)):
        name = 'slice'+str(i)+'.tiff'
        print(name)
        if name.endswith('.tiff')==False:
            pass
        else:
            img_exp = cv.imread(stack_directory+'/'+name,cv.IMREAD_GRAYSCALE)
            i=i+1
            img_bin[i] = img_exp
            
    return img_bin


directory = 'TestData/pores-hi-res-crop/'
#base.ExpProcess(directory)
```
