```python
def gyration_moments_2(G, weighted=True, sampling=1):
#Serial implementation
    Ax=0
    Ay=0
    ratio1 = 0
    ratio2 = 0
    ratio3 = 0
    node_count = np.asarray(list(range(G.vcount())))
    mask = np.random.rand(G.vcount()) > (1-sampling)
    trimmed_node_count = node_count[mask]
    for i in trimmed_node_count:
        for j in trimmed_node_count:
            if i >= j:    #Symetric matrix
                continue
            path = G.get_shortest_paths(i,to=j)
            Ax_term = 0
            Ay_term = 0
            Az_term = 0
            ratio_term1 = 0
            ratio_term2 = 0
            ratio_term3 = 0
            for hop_s,hop_t in zip(path[0][0:-1],path[0][1::]):
                if weighted:
                    weight = G.es[G.get_eid(hop_s,hop_t)]['weight']
                else:
                    weight = 1
                Ax_term = Ax_term + ((weight*(int(G.vs[hop_s]['o'][0])-int(G.vs[hop_t]['o'][0]))))
                Ay_term = Ay_term + ((weight*(int(G.vs[hop_s]['o'][1])-int(G.vs[hop_t]['o'][1]))))
                Az_term = Az_term + ((weight*(int(G.vs[hop_s]['o'][2])-int(G.vs[hop_t]['o'][2]))))
                try:
                    ratio_term1 = Ax_term/Ay_term
                    ratio_term2 = Ax_term/Az_term
                    ratio_term3 = Ay_term/Az_term
                except ZeroDivisionError:
                    pass
            ratio1 = ratio1 + ratio_term1
            ratio2 = ratio2 + ratio_term2
            ratio3 = ratio3 + ratio_term3

    print(ratio1,ratio2,ratio3)
    
def gyration_moments_3(G, sampling=1):
#Serial implementation
    Ax=0
    Ay=0
    Az=0
    node_count = np.asarray(list(range(G.vcount())))
    mask = np.random.rand(G.vcount()) > (1-sampling)
    trimmed_node_count = node_count[mask]
    for i in trimmed_node_count:
        for j in trimmed_node_count:
            if i >= j:    #Symetric matrix
                continue
            path = G.get_shortest_paths(i,to=j)
            Ax_term = 0
            Ay_term = 0
            Az_term = 0
            for hop_s,hop_t in zip(path[0][0:-1],path[0][1::]):
                weight = 1
                Ax_term = Ax_term + ((weight*(int(G.vs[hop_s]['o'][0])-int(G.vs[hop_t]['o'][0]))))
                Ay_term = Ay_term + ((weight*(int(G.vs[hop_s]['o'][1])-int(G.vs[hop_t]['o'][1]))))
                Az_term = Az_term + ((weight*(int(G.vs[hop_s]['o'][2])-int(G.vs[hop_t]['o'][2]))))
            Ax = Ax + (Ax_term)
            Ay = Ay + (Ay_term)
            Az = Az + (Az_term)

    print(Ax,Ay,Az) 
```
